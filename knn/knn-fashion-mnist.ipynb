```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_train = np.array(pd.read_csv("fashionmnist/fashion-mnist_train.csv"))
data_test = np.array(pd.read_csv("fashionmnist/fashion-mnist_test.csv"))
```

## Divida o conjunto em treino, validação e teste (70, 10, 20)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train = data_train[:, 1:]
y_train = np.ravel(data_train[:, 0:1])
```

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(X_train, y_train, test_size=0.2, train_size=0.7,random_state=42)
```

```python
(trainData, valData, trainLabels, valLabels) = train_test_split(trainData, trainLabels, test_size=0.1, random_state=84)
```

```python
print("Training data points: {}".format(len(trainLabels)))
print("Validation data points: {}".format(len(valLabels)))
print("Testing data points: {}".format(len(testLabels)))
```

```python
accuracies = []
```

```python
for k in range(1, 30, 1):
  model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
  model.fit(trainData, trainLabels)

  score = model.score(valData, valLabels)
  print("K = %d, accuracu = %.2f%%" % (k, score * 100))
  accuracies.append(score)
```

```python
kvals = range(1,30,1)
plt.plot(kvals, accuracies)
```
